7. ¿Hay valores atípicos de la variable `Weight`? Confírmalo con un gráfico de boxplot. ¿Cuántos son? Almacénalos en un dataframe independiente y elimínalos del conjunto original.

```python

# Create a boxplot of the 'Weight' column
sns.boxplot(x=response_df["Weight"])
plt.show()

# Identify outlier values based on the interquartile range (IQR) method
Q1 = response_df["Weight"].quantile(0.25)
Q3 = response_df["Weight"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Select rows with outlier values
outlier_rows = response_df[(response_df["Weight"] < lower_bound) | (response_df["Weight"] > upper_bound)]

# Count the number of outlier values
outlier_count = len(outlier_rows)
outlier_df = outlier_rows.copy()

# Remove outlier rows from the original dataframe
response_df = response_df[~((response_df["Weight"] < lower_bound) | (response_df["Weight"] > upper_bound))]
print(f"Number of outlier values in 'Weight': {outlier_count}")
```

8. Repite el mismo proceso de valores atípicos con la variable `Height`.

```python
# Create a boxplot of the 'Height' column
sns.boxplot(x=response_df["Height"])
plt.show()

# Identify outlier values based on the interquartile range (IQR) method
Q1 = response_df["Height"].quantile(0.25)
Q3 = response_df["Height"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Select rows with outlier values
outlier_rows = response_df[(response_df["Height"] < lower_bound) | (response_df["Height"] > upper_bound)]

# Count the number of outlier values
outlier_count = len(outlier_rows)
outlier_df = outlier_rows.copy()

# Remove outlier rows from the original dataframe
response_df = response_df[~((response_df["Height"] < lower_bound) | (response_df["Height"] > upper_bound))]
print(f"Number of outlier values in 'Height': {outlier_count}")
```
